### Shift Code

Pseduo code

- Create option menu
    1. Code a message
    2. Decode a message
    3. Quit
- Read users input
    Generate error if input is invalid

- Option 1
- Ask user for message
- Read message
- Ask user for code_number
- Read code_number
- Move each letter forward code_number spaces
- Write to new coded_message
- Print out new coded_message along with code_number

- Option 2
- Ask user for coded_message
- Read coded_message
- Ask user for code_number
- Read code_number
- Move each letter backward code_number spaces
- Write to new message
- Print out message 

- Option 3
- End program

```python
def code_message():
    message = list((input('Enter your message:')).lower())
    code_number = int((input('Enter your code number:')))
    for c, letter in enumerate(message):
        message[c] = ord(message[c])
    for c ,letter in enumerate(message):
        message[c] = message[c]+code_number
        if message[c] >= 123:
            message[c] = (message[c] - 123) + 97
    for c, letter in enumerate(message):
        message[c] = chr(message[c])
    message = ''.join(message)
    return print('Your coded message is:',message,'\nUsing code number:',code_number)

def decode_message():
    coded_message = list((input('Enter your coded message:')).lower())
    code_number = int((input('Enter your code number:')))
    for c, letter in enumerate(coded_message):
        coded_message[c] = ord(coded_message[c])
    for c ,letter in enumerate(coded_message):
        coded_message[c] = coded_message[c]-code_number
        if coded_message[c] <= 96:
            coded_message[c] = 123 - (97 - coded_message[c])
    for c, letter in enumerate(coded_message):
        coded_message[c] = chr(coded_message[c])
    coded_message = ''.join(coded_message)
    return print('Your original message is:',coded_message,'\nUsing code number:',code_number)


print('1. Code a message\n2. Decode a message\n3. Quit')
select = int(input('Enter your selection:'))


if select == 1:
    code_message()

if select == 2:
    decode_message()

if select == 3:
    quit
```
